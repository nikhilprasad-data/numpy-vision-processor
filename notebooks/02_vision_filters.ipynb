```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_1 = plt.imread('../image/processed/processed_bike_001.jpg')
plt.imshow(image_1)
plt.axis('off')
```

```python
image_2 = plt.imread('../image/processed/processed_bike_002.jpg')
plt.imshow(image_2)
plt.axis('off')
```

```python
image_3 = plt.imread('../image/processed/processed_bike_003.jpg')
plt.imshow(image_3)
plt.axis('off')
```

```python
red_slice_image_1 = image_1[:, :, 0] 
red_slice_image_1.shape
```

```python
plt.imshow(red_slice_image_1)
plt.axis('off')
```

```python
bw_image_1 = np.mean(image_1, axis= 2)
bw_image_1.shape
```

```python
plt.imshow(bw_image_1, cmap= 'gray')
plt.axis('off')
```

```python
bw_image_1[bw_image_1 < 50] = 0
plt.imshow(bw_image_1, cmap= 'gray')
plt.axis('off')
```

```python
bw_image_3 = np.mean(image_3, axis= 2)
plt.imshow(bw_image_3, cmap= 'gray')
plt.axis('off')
```

```python
bw_image_3[bw_image_3 < 50] = 0
plt.imshow(bw_image_3, cmap= 'gray')
plt.axis('off')
```

```python
hstack_image_1_3 = np.hstack((bw_image_1, bw_image_3))
plt.imshow(hstack_image_1_3, cmap= 'gray')
plt.axis('off')
```

```python
hstack_image_1_3.shape
```

```python
bw_image_2 = np.mean(image_2, axis=2)
plt.imshow(bw_image_2, cmap= 'gray')
plt.axis('off')
```

```python
bw_image_2[bw_image_2 < 50] = 0
plt.imshow(bw_image_2, cmap= 'gray')
plt.axis('off')
```

```python
xray_image_2 = 255 - bw_image_2
plt.imshow(xray_image_2, cmap= 'gray')
plt.axis('off')
```

```python
hstack_image_2_2 = np.hstack((xray_image_2, xray_image_2))
plt.imshow(hstack_image_2_2, cmap= 'gray')
plt.axis('off')
```

```python
vstack = np.vstack((hstack_image_2_2, hstack_image_1_3))
plt.imshow(vstack, cmap= 'gray')
plt.axis('off') 
```

```python
bw_image_1_float = bw_image_1.astype(float)
plt.imshow(bw_image_1_float, cmap= 'gray')
plt.axis('off')
```

```python
bw_image_1_float_shifted_up = bw_image_1_float[1:, 1:]
plt.imshow(bw_image_1_float_shifted_up, cmap= 'grey')
plt.axis('off')
```

```python
bw_image_1_float_shifted_down = bw_image_1_float[:-1,:-1]
plt.imshow(bw_image_1_float_shifted_down, cmap= 'grey')
plt.axis('off')
```

```python
gradient_image_1 = np.abs(bw_image_1_float_shifted_up - bw_image_1_float_shifted_down)
plt.imshow(gradient_image_1, cmap= 'gray')
plt.axis('off')
```
