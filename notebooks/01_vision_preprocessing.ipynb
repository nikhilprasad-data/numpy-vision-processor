```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading the raw JPG image from local directory into a 3D NumPy array
raw_image_1 = plt.imread('../image/raw/raw_bike_001.jpg')
plt.imshow(raw_image_1)
plt.axis('off')
```

```python
raw_image_1.shape
```

```python
# Cropping 100 pixels from top and bottom using matrix slicing to standardize height to 1080px
cropped_raw_image_1 = raw_image_1[100:1180, :]
plt.imshow(cropped_raw_image_1)
plt.axis('off')
```

```python
cropped_raw_image_1.shape
```

```python
raw_image_2 = plt.imread('../image/raw/raw_bike_002.jpg')
plt.imshow(raw_image_2)
plt.axis('off')
```
